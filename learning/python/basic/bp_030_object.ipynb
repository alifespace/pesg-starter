```python
class Student:
  def __init__(self, name, sex, age, height):
    self.name = name
    self.sex = sex
    self.age = age
    self.height = height

  def listen(self):
    print('listen...')

  def write(self):
    print('write...')

  def intro(self):
    print(self.name, self.sex, self.age, self.height)

stu1 = Student('张三', '男', 16, 167)

stu1.intro()
stu1.listen()
print(stu1.age)

stu1.age = 18
print(stu1.age)

# 动态给对象添加属性
stu1.pro1 = '新增属性'
print(stu1.pro1)
print(stu1.__dict__)

del stu1.sex
print(stu1.__dict__)

```
