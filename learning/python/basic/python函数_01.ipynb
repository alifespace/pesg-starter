```python
# sorted(): 排序
number = [5, 2, 8, 1, 10]
demo = sorted(number)
demo1 = sorted(number, reverse=True)

print("Original list:", number)
print("Sorted list:", demo)
print("Sorted list in descending order:", demo1)
print('\n')

str = ["banana", "apple", "cherry", "date"]
demo2 = sorted(str)
demo3 = sorted(str, key=len)
print("Original list of strings:", str)
print("Sorted list of strings:", demo2)
print("Sorted list of strings by length:", demo3)
```

```python
# zip(): 压缩, 将多个可迭代对象打包成一个元组列表
# zip()生成的是惰性迭代器，用一次就失效了。所以如果需要多次使用，应该将其转换为列表或元组。

name = ["Alice", "Bob", "Charlie"]
age = [25, 30, 35]
gender = ["Female", "Male", "Male"]

zipped = zip(name, age, gender)
zipped_list = list(zipped)  # 转换为列表以供多次使用

print('zipped type:', type(zipped))
print("Zipped result:", zipped_list)

elm1, elm2, elm3 = zip(*zipped_list)  # 解压缩
print("Unzipped result:", elm1, elm2, elm3)
```
