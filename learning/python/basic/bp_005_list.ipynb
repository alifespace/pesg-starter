```python
names = ['张三', '李四', '王五', '郑六1']

print(names)
print(names[-2])

names.append('田七1')
print(names)

names.insert(2, 'xxx')   # 在 index 为2的位置插入数据
print(names)

# 删除列表的元素
del names[1]
print(names)

n1 = names.pop(-2)
print(n1, names)

names.remove('xxx')   # 只删除第一个找到的元素
print(names)
```

```python
names = ['cc', 'bb', 'aa', 'dd']

tmp1 = sorted(names)
print(tmp1, names)

names.sort()
print(names)

names.reverse()
print(names)

n1 = len(names)
print(n1)

# print(names[4])
print(n1)
```

1. 添加：.append(), .insert()
2. 删除：del, .pop(), .remove()
3. 其他：sorted, .sort(), len
