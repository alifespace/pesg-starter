```python
import sqlite3
import os
import pandas as pd

# 数据库路径（与 notebook 同目录）
db_path = os.path.join(os.path.dirname("__file__"), "longlist0.db")

# 建立连接
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# 验证连接成功，并查看所有表
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cursor.fetchall()
print(f"✅ 已连接到 {db_path}")
print(f"📋 包含 {len(tables)} 张表：")
for t in tables:
    print(f"   - {t[0]}")
```

```python
# 查询示例
df = pd.read_sql_query("SELECT * FROM longlist LIMIT 10", conn)
df
```

```python
# 数据库适合存储大量结构化数据，尤其是关系型数据。它提供了高效的数据存储和查询能力，支持事务处理和数据完整性约束。
# 数据库优势包括：
# 1. 数据一致性：通过事务和约束，确保数据的完整性和一致性。
# 2. 高效查询：使用索引和优化的查询引擎，能够快速检索大量数据。
# 3. 并发访问：支持多用户同时访问和操作数据，适合团队协作。
# 4. 数据安全：提供权限管理和数据加密，保护敏感信息。
# 5. 可扩展性：适用于从小型应用到大型企业级系统的各种场景。
# CRUD 操作是数据库的基本操作，包括
# 创建（Create）、读取（Read）、更新（Update）和删除（Delete）数据。通过这些操作，用户可以对数据库中的数据进行管理和维护。

# SQL: Structured Query Language（结构化查询语言），是一种用于管理和操作关系型数据库的标准语言。
# SQL 提供了丰富的语法和功能，用于执行数据查询、插入、更新、删除以及数据库结构定义等操作。
# 通过 SQL，用户可以高效地与数据库进行交互，实现数据的存储、检索和分析。

# df = pd.read_sql_query("SELECT * FROM books LIMIT 10", conn)
# df

# SQL是一种声明式语言，用户只需描述想要的数据结果，而不需要指定具体的操作步骤。SQL的核心功能包括：
df = pd.read_sql_query('SELECT title FROM longlist LIMIT 10', conn)
df

df = pd.read_sql_query('SELECT title from longlist LIMIT 5', conn)
df
```

```python
df = pd.read_sql_query('SELECT title, author FROM longlist WHERE year = 2023', conn)
df
```

```python
# SQL中不等于有两种写法：<> 和 !=，它们的作用相同，都是用于表示不等于的条件。
# 不同数据库系统可能对这两种写法的支持有所不同，但在大多数情况下，它们是可以互换使用的。

df = pd.read_sql_query('SELECT title, format FROM longlist WHERE format != \'Hardcover\'', conn)
df
```

```python
# NOT: 用于在 SQL 查询中表示逻辑非操作，通常用于条件表达式中，以排除满足特定条件的记录。

df = pd.read_sql_query('SELECT title, author FROM longlist WHERE NOT format = \'Hardcover\'', conn)
df
```

```python
# 逻辑运算符 AND 和 OR 用于在 SQL 查询中组合多个条件，以实现更复杂的筛选逻辑。

df = pd.read_sql_query('SELECT title, author FROM longlist WHERE year = 2022 OR year = 2023', conn)
df
```

```python
df = pd.read_sql_query('SELECT title, author, format FROM longlist WHERE year = 2022 OR year = 2023 AND format = \'Paperback\'', conn)
df
```

```python
# NULL: 用于表示缺失或未知的值。在 SQL 查询中，可以使用 IS NULL 或 IS NOT NULL 来筛选包含或不包含 NULL 值的记录。
df = pd.read_sql_query('SELECT title, author, translator FROM longlist WHERE translator IS NULL', conn)
df
```

```python
df = pd.read_sql_query('SELECT title, author, translator FROM longlist WHERE translator IS NOT NULL', conn)
df
```

```python
#LIKE: 用于在 SQL 查询中进行模糊匹配，通常与通配符（如 % 和 _）结合使用，以查找符合特定模式的记录。
# %: 表示任意数量的字符（包括零个字符）。
# _: 表示单个字符。
df = pd.read_sql_query('SELECT title FROM longlist WHERE title LIKE \'%love%\'', conn)
df
```

```python
df = pd.read_sql_query('SELECT title, author FROM longlist WHERE title LIKE \'the %\'', conn)
df
```

```python
df = pd.read_sql_query('SELECT title, author FROM longlist WHERE title LIKE \'P_re\'', conn)
df
```

```python
df = pd.read_sql_query('SELECT title, year FROM longlist WHERE year >= 2020 AND year <= 2022', conn)
# df = pd.read_sql_query('SELECT title, year FROM longlist WHERE year BETWEEN 2020 AND 2022', conn)
df
```

```python
df = pd.read_sql_query('SELECT title, rating FROM longlist WHERE rating BETWEEN 4 AND 5', conn)
df
```

```python
# ASC 和 DESC 是 SQL 中用于指定排序顺序的关键字。ASC 表示升序（从小到大），而 DESC 表示降序（从大到小）。
# 在 ORDER BY 子句中使用这两个关键字，可以对查询结果进行排序。

df = pd.read_sql_query('SELECT title, author, rating, votes FROM longlist ORDER BY rating DESC , votes DESC LIMIT 10', conn)
df
```

```python
df = pd.read_sql_query('SELECT ROUND(AVG(rating), 2) AS average_rating FROM longlist', conn)
df
```

```python
df = pd.read_sql_query('SELECT MAX(rating) AS highest_rating, MIN(rating) AS lowest_rating FROM longlist', conn)
df
```

```python
df = pd.read_sql_query('SELECT SUM(votes) AS total_votes FROM longlist', conn)
df
```

```python
df = pd.read_sql_query('SELECT COUNT(*) AS total_books FROM longlist', conn)
df
```

```python
# MAX 按照字母顺序比较字符串，返回字母顺序最大的字符串。MIN 按照字母顺序比较字符串，返回字母顺序最小的字符串。

df = pd.read_sql_query('SELECT MAX(title) AS longest_title FROM longlist', conn)
df
```

```python
# COUNT(DISTINCT column_name) 用于计算指定列中不同值的数量。它会忽略重复的值，只统计唯一的值，从而得到该列中不同值的总数。

df = pd.read_sql_query('SELECT COUNT(DISTINCT publisher) AS total_publishers FROM longlist', conn)
df
```
