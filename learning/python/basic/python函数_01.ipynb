```python
# sorted(): 排序
number = [5, 2, 8, 1, 10]
demo = sorted(number)
demo1 = sorted(number, reverse=True)

print("Original list:", number)
print("Sorted list:", demo)
print("Sorted list in descending order:", demo1)
print('\n')

str = ["banana", "apple", "cherry", "date"]
demo2 = sorted(str)
demo3 = sorted(str, key=len)
print("Original list of strings:", str)
print("Sorted list of strings:", demo2)
print("Sorted list of strings by length:", demo3)
```

```python
# zip(): 压缩, 将多个可迭代对象打包成一个元组列表
# zip()生成的是惰性迭代器，用一次就失效了。所以如果需要多次使用，应该将其转换为列表或元组。

name = ["Alice", "Bob", "Charlie"]
age = [25, 30, 35]
gender = ["Female", "Male", "Male"]

zipped = zip(name, age, gender)
zipped_list = list(zipped)  # 转换为列表以供多次使用

print('zipped type:', type(zipped))
print("Zipped result:", zipped_list)

elm1, elm2, elm3 = zip(*zipped_list)  # 解压缩
print("Unzipped result:", elm1, elm2, elm3)
```

```python
# map(): 映射, 将函数应用于可迭代对象的每个元素

def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
squared_list = list(squared_numbers)  # 转换为列表以供多次使用
print("Original numbers:", numbers)
print("Squared numbers:", squared_list)
```

```python
# filter(): 过滤, 根据条件筛选可迭代对象的元素

def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(is_even, numbers)
even_list = list(even_numbers)  # 转换为列表以供多次使用
print("Original numbers:", numbers)
print("Even numbers:", even_list)
```

```python
# reduce(): 归约, 将可迭代对象的元素累积为单个值
from functools import reduce

def multiply(x, y):
    return x * y

def add(x, y):
    return x + y

numbers = [1, 2, 3, 4, 5]
product = reduce(multiply, numbers)
print("Original numbers:", numbers)
print("Product of numbers:", product)
sum = reduce(add, numbers)
print("Sum of numbers:", sum)
```
