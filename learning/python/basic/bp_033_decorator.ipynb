```python
class Student:
  def __init__(self, age):
    self.__age = age  ## __age: 私有属性，不能直接修改

  @property  # 装饰器，需要先定义
  def age(self):
    return self.__age

  @age.setter
  def age(self, age):
    self.__age = age

  @age.deleter
  def age(self):
    del self.__age
    print('删除了age属性')

stu1 = Student(10)
stu1.age = 20
print('stu1:', stu1.age)
```
