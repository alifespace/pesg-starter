```python
class A:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def foo1(self, x):
    print('foo1')
    pass

  def foo2(self, x):
    print('foo2')
    pass

class B(A):
  def foo1(self, x):
    if x == 3:
      print(f'我叫{self.name}')
    else:
      super().foo1(x)

b = B('张三', 12)
b.foo1(2)
b.foo2(2)
```
