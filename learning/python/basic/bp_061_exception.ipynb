```python
try:
  raise Exception('Invalid')
except Exception as e:
  print('处理错误：', e)
```

```python
try:
  raise ZeroDivisionError('除零错误')
except ZeroDivisionError as e:
  print('处理错误: ', e)

try:
  n1 = 7 / 0
except Exception as e:
  print('处理错误：', e)
```

```python
import pathlib
try:
  pathlib.Path('ab.txt').read_text()
except Exception as e:
  print('处理错误：', e)
```

```python
try:
  n1 = 'abcd' + 123
except Exception as e:
  print('处理错误：', e)
```

```python
try:
  age = int(input('请输入年龄：'))
except Exception as e:
  print('处理错误：', e)
```

```python
try:
  dict1 = {'age': 5}
  print(dict1['age'])
except KeyError as e:
  print('处理特定KeyError错误：', e)
except Exception as e:
  print('处理错误：', e)
else:
  print('没有错误，正常执行。')
finally:
  print('不管成功还是失败，都会执行这里')
```
