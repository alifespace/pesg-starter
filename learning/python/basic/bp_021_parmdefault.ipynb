```python
def say(name, friends=[], parm1='1'):
  friends.append(name)
  print(friends, parm1)

# say('李四', ['张三'], '2')
say('张三')
say('李四', ['abc'])
say('李四')
```

```python
def say(name, friends=None):
  if friends is None:
    friends = []
  friends.append(name)
  print(friends)

say('张三')
say('李四')
```
